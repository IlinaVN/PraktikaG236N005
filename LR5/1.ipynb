```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\Администратор\\Documents\\M3023622\\ИльинаЖирков\\LR5\\data.csv', sep=',')
```

```python
df['days_employed'] = abs(df['days_employed'])
```

```python
df['days_employed'] = df['days_employed'].fillna(df['days_employed'].median())
```

```python
df['total_income'] = df['total_income'].fillna(0)
```

```python
df['total_income'] = df['total_income'].astype(int)
```

```python
df['education'] = df['education'].str.lower()
```

```python
df = df.drop_duplicates(keep=False)
```

```python
income_cat = len(df) * [0]
```

```python
for i in range(len(df)):
    if df['total_income'].loc[df.index[i]] < 30000:
        income_cat[i] = 'E'
    elif df['total_income'].loc[df.index[i]] < 50000:
        income_cat[i] = 'D'
    elif df['total_income'].loc[df.index[i]] < 200000:
        income_cat[i] = 'C'
    elif df['total_income'].loc[df.index[i]] < 1000000:
        income_cat[i] = 'B'
    else:
        income_cat[i] = 'A'
```

```python
df.insert(11,'total_income_category', income_cat)
```

```python
def purpose_cat(df):
    if 'автомоб' in df:
        return 'операции с автомобилем'
    if 'жиль' in df:
        return 'операции с недвижимостью'
    if 'недвиж' in df:
        return 'операции с недвижимостью'
    if 'свадь' in df:
        return 'проведение свадьбы'
    if 'образов' in df:
        return 'получение образования'
```

```python
df['purpose_category'] = df['purpose'].apply(purpose_cat)
```
